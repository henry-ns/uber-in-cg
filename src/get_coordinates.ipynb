```python
import geocoder as geo # https://geocoder.readthedocs.io/
import requests as req
import numpy as np
import pandas as pd
import json
```

```python
bairros = pd.read_csv('../res/bairros.csv')['Bairros'].tolist()
```

```python
locates = []
for locate in bairros: 
    g = geo.arcgis(locate + ', Campina Grande, PB')
    
    l = {
        'address': g.address,
        'bbox': g.json['bbox'],
        'latlng': g.latlng
    }
    
    if l not in locates:
        locates.append(l)
        print(g.address, ' - ', g.latlng)
```

```python
file = open('../res/coordinates.json', 'w+')
file.write(json.dumps(locates))
```
