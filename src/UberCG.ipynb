```python
import pandas as pd
```

        No codigo abaixo está sendo feito a importação dos dados que estão armazenados num arquivo CSV. Em seguida deixo mais legivel o horario deixando no formato YY-MM-DD HH:mm e definindo que o index do DataFrame é justamente o horario.

```python
df = pd.read_csv("../res/bairros_estimativas.csv")
for i, j in zip(df['horario'], range(0, df.shape[0])):
    df['horario'][j] = i[2:16]
n_column = []
for i in df.columns:
    n_column.append(i.replace(', Campina Grande, Paraíba', ''))
df.columns = n_column
df = df.set_index("horario") 
```

        Como é possivel ver abaixo, cada linha do DataFrame df é composto pelo horario no qual a requisição foi feita na API da Uber e o tempo ,em segundos, na qual um UberX demoraria para chegar nas coordenadas, no df é mostrado apenas o nome da localização para ficar de vizualização mais fácil.

```python
df
```

            No grafico de barras horizontais abaixo é mostrado, em ordem descedente, o tempo de espera em cada ponto no horario de 10:18 do dia 15/12/18. Para ver para os outros horarios mude o valor de x

```python
x = 14
df.iloc[x].sort_values().plot.barh(figsize=(10,20))
```

            No grafico de linhas abaixo é mostrado, para o bairro do Alto Branco, o tempo de espera ao longo do dia. Lembrando que os dados foram coletados entre 09:47 e 18:03 do dia 15/12/18.

```python
df['Alto Branco'].plot(figsize=(15,10), x='horario')
```

            O boxplot abaixo representa informações sobre o horario de 09:47 de todos os pontos onde foram coletados dados.

```python
print(df.iloc[0].describe())
df.iloc[0].plot.box()
```

            Já o boxplot abaixo mostra as informações somente do bairro Alto Branco.

```python
print(df['Alto Branco'].describe())
df['Alto Branco'].plot.box()
```

            As informações abaixo informam, respectivamente, a contagem, média, desvio padrão, valor minimo, primeiro quartil, mediana, terceiro quartil e valor maximo de cada horario daquele dia.

```python
dd = []
for i in range(df.shape[0]):
    dd.append(df.iloc[i].describe())
pd.DataFrame(dd)
```

            Abaixo temos a mesma analise porém para os bairros

```python
dd = []
for i in range(df.shape[1]):
    dd.append(df.transpose().iloc[i].describe())
pd.DataFrame(dd)
```
