```python
import pandas as pd
```

```python
df = pd.read_csv("bairros_estimativas.csv")
for i, j in zip(df['horario'], range(0, df.shape[0])):
    df['horario'][j] = i[11:16]
n_column = []
for i in df.columns:
    n_column.append(i.replace(', Campina Grande, Paraíba', ''))
df.columns = n_column
df = df.set_index("horario")
```

```python
df_mean = df.reset_index().drop('horario', axis=1)
df_mean
mean = {}
for column in df_mean.columns:
    m = df_mean[column].mean()
    mean[column] = m
df_mean = pd.DataFrame(data=mean, index=['media'])
print(df_mean.transpose().nlargest(10, 'media'))
df_mean
```
